### Drawing on Images

```python
#Creating a blank image
blank_img=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
#Drawing rectange over image, pt1 is the first vertex of rectange(top-left), pt2 is the second vertex of the rectangle(bottom-right)
#In pt1, first point is on x axis and second on y
#color code is RGB,as the image is in RGB color code, otherwise BGR
#It will not return anything, just changing in the original image.
cv2.rectangle(blank_img,pt1=(380,10),pt2=(500,200),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
#Drawing a square in the middle of the image(thickness can be set with the color also provided)
cv2.rectangle(blank_img,pt1=(200,200),pt2=(400,400),color=(255,0,0),thickness=6);
```

```python
plt.imshow(blank_img)
```

```python
#Drawing a circle,  center is the point of the center of circle,   radius is how much we need to expand it(in px)
cv2.circle(blank_img,center=(100,50),radius=50,color=(0,0,255),thickness=6);
```

```python
plt.imshow(blank_img)
```

```python
#Drawing a filled circle, it just takes thickness as -1
cv2.circle(blank_img,center=(50,450),radius=50,color=(0,0,255),thickness=-1);
plt.imshow(blank_img)
```

```python
#Drawing a line
cv2.line(blank_img,pt1=(15,380),pt2=(15,200),color=(0,255,0),thickness=4);
plt.imshow(blank_img)
```

```python
#Drawing Text, we can only select font from cv2,'org' parameter takes bottom left coordinate
#FontScale decides the size of the font, and color is given as white(when all are 255)
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text="Image",org=(80,250),fontFace=font,fontScale=3,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA);
plt.imshow(blank_img)
```

### Creating customized polygon

```python
#creating a blank image
#dtype is used as we do not want it to be in float, it should be in interger
blank_img_poly=np.zeros(shape=(1320,1920,3),dtype=np.int16)
```

```python
plt.imshow(blank_img_poly)
```

```python
#All these contains the coordinates
vertices=np.array([ [100,300],[200,200],[400,300],[200,400] ],dtype=np.int32)
```

```python
vertices.shape
```

```python
#OpenCV wants these points in 3 dimensions because of color channels(we should always write these numbers for reshaping)
#(keeping the same information)
pts=vertices.reshape((-1,1,2))
```

```python
#It has just added another dimension as 1, keeping the same information
pts.shape
```

```python
#Creating a polygon, [pts] we need to pass as a list as it contains points(above it contains coordinates)
#isClosed ensures the polygon must be closed or not
cv2.polylines(blank_img_poly,[pts],isClosed=True,color=(120,125,0),thickness=10);
plt.imshow(blank_img_poly)
```

### Direct Drawing images using Mouse
We can use CallBacks to connect images to event functions with OpenCV, This allows you to directly interact with images.

It is divied into parts:-
* Connecting Callback Functions
* Adding functionality through Event Choices
* Dragging the Mouse for functionality

We must keep it in a single cell or single python script file

**Drawing Circle using Mouse**

```python
#0xFF==27 means that if anyone presses Esc key
#dtype=int8 helps to make the background gray
#color is swapped to BGR
```

```python
import cv2
import numpy as np

#Function(these parameters will be automatically filled by the SetMouseCallback)
#We do not need to pass center points (x,y) as it will be taken by mouse
def draw_circle(event,x,y,flags,param):
    if event==cv2.EVENT_LBUTTONDOWN: #if the left mouse button is pressed, then draw a circle
        cv2.circle(img,(x,y),radius=100,color=(0,255,0),thickness=-1)
    
    elif event==cv2.EVENT_RBUTTONDOWN:#if the right mouse button is pressed, then draw a different circle
        cv2.circle(img,(x,y),radius=60,color=(255,125,0),thickness=6)

#Helps to connect above function with the image callback
cv2.namedWindow(winname='my_drawing')
cv2.setMouseCallback('my_drawing',draw_circle)


img=np.zeros((512,512,3),dtype=np.int8)

while True:
    cv2.imshow("my_drawing",img)
    
    if cv2.waitKey(20) & 0xFF==27:
        break

cv2.destroyAllWindows()
```

**Drawing with Mouse Freehand**

```python
#drawing=false, it will be true when mouse button is down and again to false when mouse button is up.
#x,y keeps the track where the mouse is currently located
```

```python
import cv2
import numpy as np

#Variables
drawing=False 
ix=iy=-1,-1


#Function(these parameters will be automatically filled by the SetMouseCallback)
def draw_rectangle(event,x,y,flags,param):
    global drawing,ix,iy
    
    if event==cv2.EVENT_LBUTTONDOWN: 
        drawing=True #as the mouse button is pressed/down
        ix,iy=x,y #will change the value to the place where the mouse was located
        
    
    elif event==cv2.EVENT_MOUSEMOVE: #movement of mouse
        if drawing==True:
            cv2.rectangle(img,pt1=(ix,iy),pt2=(x,y),color=(0,255,0),thickness=-1)
    
    elif event==cv2.EVENT_LBUTTONUP:
        drawing=False
        cv2.rectangle(img,pt1=(ix,iy),pt2=(x,y),color=(0,255,0),thickness=-1)

#Helps to connect above function with the image callback
cv2.namedWindow(winname='my_drawing')
cv2.setMouseCallback('my_drawing',draw_rectangle)


img=np.zeros((512,512,3),dtype=np.int8)

while True:
    cv2.imshow("my_drawing",img)
    
    if cv2.waitKey(20) & 0xFF==27:
        break

cv2.destroyAllWindows()
```
